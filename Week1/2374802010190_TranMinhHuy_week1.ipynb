```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col="STT")
df.head(10)
```

```python
df['DT'].fillna(0, inplace=True)
df
```

```python
#Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean.
so_luong_thieu = df['T1'].isnull().sum()
print(f"Số lượng giá trị thiếu của T1: {so_luong_thieu}")
mean_t1 = df['T1'].mean()
df['T1'] = df['T1'].fillna(mean_t1)
df.head()
```

```python
#Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.
cols_score = []
mon_hoc = ['T', 'L', 'H', 'S', 'V', 'X', 'D', 'N']
for mon in mon_hoc:
    for i in range(1, 7):
        cols_score.append(f'{mon}{i}')
cols_score.extend(['DH1', 'DH2', 'DH3'])
for col in cols_score:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())
df.head()
```

```python
#Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và
12.
for i in ['1', '2', '3']:
    df[f'TBM{i}'] = (df[f'T{i}']*2 + df[f'L{i}'] + df[f'H{i}'] + df[f'S{i}'] +
                     df[f'V{i}']*2 + df[f'X{i}'] + df[f'D{i}'] + df[f'N{i}']) / 10
df.head()
```

```python
#Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp
10, 11, 12
def xep_loai(x):
    if x < 5.0: return 'Y'
    elif x < 6.5: return 'TB'
    elif x < 8.0: return 'K'
    elif x < 9.0: return 'G'
    else: return 'XS'

for i in ['1', '2', '3']:
    df[f'XL{i}'] = df[f'TBM{i}'].apply(xep_loai)
df.head()
```

```python
#Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp
10, 11
#và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương
#pháp Min-Max Normalization

for i in ['1', '2', '3']:
    df[f'US_TBM{i}'] = df[f'TBM{i}'] / 10 * 4
df.head()
```

```python
#Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và
#rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3

def xet_tuyen(row):
    khoi = row['KT']
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']
    if khoi in ['A', 'A1']:
        diem = (dh1*2 + dh2 + dh3) / 4
    elif khoi == 'B':
        diem = (dh1 + dh2*2 + dh3) / 4
    else:
        diem = (dh1 + dh2 + dh3) / 3
    return 1 if diem >= 5 else 0

df['KQXT'] = df.apply(xet_tuyen, axis=1)
df.head()
```

```python
#lưu file
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False)
```
